```python
import time
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
import undetected_chromedriver as uc
from selenium.webdriver.common.action_chains import ActionChains

class ExplainPaper:
    def __init__(self):
        self.driver = None

    def open_chrome_browser(self):
        self.driver = uc.Chrome(execfile="/home/dibya/Soft/anaconda3/lib/python3.9/site-packages/chromedriver_autoinstaller/109/chromedriver")
        self.driver.get("https://www.explainpaper.com")

    def login(self, email, password):
        time.sleep(5)
        btns = self.driver.find_elements(By.TAG_NAME, 'button')
        button = WebDriverWait(self.driver, 10).until(EC.element_to_be_clickable(btns[0]))
        button.click()
        time.sleep(5)
        btns = self.driver.find_elements(By.TAG_NAME, 'button')
        button = WebDriverWait(self.driver, 10).until(EC.element_to_be_clickable(btns[2]))
        button.click()
        time.sleep(5)
        email_input = self.driver.find_element(By.XPATH, '//*[@id="identifierId"]')
        email_input.send_keys(email)
        self.driver.find_element(By.XPATH, '//*[@id="identifierNext"]/div/button/span').click()
        time.sleep(3)
        password_input = self.driver.find_element(By.XPATH, '//*[@id="password"]/div[1]/div/div[1]/input')
        password_input.send_keys(password)
        self.driver.find_element(By.XPATH, '//*[@id="passwordNext"]/div/button/span').click()
        time.sleep(5)

    def upload_file(self, file_path):
        time.sleep(5)
        clickable = self.driver.find_elements(By.TAG_NAME, 'button')[1]
        WebDriverWait(self.driver, 20).until(EC.element_to_be_clickable(clickable)).click()
        time.sleep(3)
        file_input = self.driver.find_element(By.XPATH, "//input[@type='file']")
        file_input.send_keys(file_path)
        time.sleep(5)
        analyze = self.driver.find_elements(By.TAG_NAME, 'button')
        file_input = WebDriverWait(self.driver, 10).until(EC.element_to_be_clickable(analyze[3]))
        file_input.click()
        time.sleep(5)
    def double_click_element(self, index):
        elem = self.driver.find_elements(By.TAG_NAME, 'span')
        actions = ActionChains(self.driver)
        actions.double_click(elem[index])
        actions.perform()
        sleep(10)

    def click_question(self, index):
        question = self.driver.find_elements(By.TAG_NAME, 'p')
        file_input = WebDriverWait(self.driver, 30).until(EC.element_to_be_clickable(question[index]))
        file_input.click()
        sleep(2)

    def ask_question(self, text):
        # Find the 'textarea' element to enter the question
        text_element = WebDriverWait(self.driver, 30).until(EC.element_to_be_clickable((By.TAG_NAME, 'textarea')))
        text_element.clear()
        # Enter the question text
        text_element.send_keys(text)
        text_element.send_keys(Keys.RETURN)
        time.sleep(10)
        # Find the answer element
        answer = WebDriverWait(self.driver, 30).until(EC.presence_of_element_located((By.ID, 'chat-panel')))
        # Split the answer into lines and return the second line
        return answer.text.split("\n")[1]
```

```python
# create an object of ExplainPaper class
ep = ExplainPaper()

# open the Chrome browser
ep.open_chrome_browser()

# login to ExplainPaper
ep.login('1999dibyam', 'Ansv@1098')
# upload a file
ep.upload_file('/home/dibya/Downloads/1ff179c3ad5c52fca11f792c3a09a134.pdf')
try :
    ep.double_click_element(30)
except:
    ep.double_click_element(31)
ep.click_question(5)
# ask a question
answer = ep.ask_question('Explain the paper in great details!')

# print the answer
print(answer)
```
